```python
#importing label encoding module
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()

#creating anew data frame from performing the chi2 analysis
df1=pd.DataFrame()

#adding all the categorical columns expect the output to new data frame
for i in categorical:
    if i!= 'Accident_severity':
        df1[i]=le.fit_transform(df[i])
```

```python
#import chi2 test
from sklearn.feature_selection import chi2
f_p_values=chi2(df1,df['Accident_severity'])
```

```python
#f_p_values will return Fscore and pvalues
f_p_values
```

```python
#for better understandings and ease of access adding them to a new dataframes
f_p_values1=pd.DataFrame({'features':df1.columns,'Fscore': f_p_values[0], 'Pvalues':f_p_values[1]})
f_p_values1
```

```python
#since we want lower pvalues we are sorting the features
f_p_values1.sort_values(by='Pvalues',ascending=True)
```

```python
#after evaluating we are removing lesser important columns and sorting to a new data frame
df2=df.drop(['Owner_of_vehicle','Type_of_vehicle','Road_surface_conditions', 'Pedestrian_movement','Casualty_severity','Educational_level','Day_of_week','Sex_of_driver','Road_allignment','Sex_of_casualty'],axis=1)
df2.head()
```

```python
#to check distinct values in each categorical columns we are sorting them to a new variable
categorical_new = df2.select_dtypes(include=['object', 'category']).columns.tolist()
print(categorical_new)
```

```python
for i in categorical_new:
    print(df2[i].value_counts())
```

```python
#get dummies
cols_to_encode = [
    'Age_band_of_driver', 'Vehicle_driver_relation', 'Driving_experience',
    'Area_accident_occured', 'Lanes_or_Medians', 'Types_of_Junction',
    'Road_surface_type', 'Light_conditions', 'Weather_conditions',
    'Type_of_collision', 'Casualty_class',
    'Age_band_of_casualty', 'Cause_of_accident'
    # 'vehicle_movement' removed
]


dummy = pd.get_dummies(df2[cols_to_encode], drop_first=True)
display(dummy.head())

```

```python
#concatinate dummy and old data frame
df3=pd.concat([df2,dummy],axis=1)
df3.head()
```

```python
#dropping dummied columns
df3.drop([
    'Age_band_of_driver', 'Vehicle_driver_relation', 'Driving_experience',
    'Area_accident_occured', 'Lanes_or_Medians', 'Types_of_Junction',
    'Road_surface_type', 'Light_conditions', 'Weather_conditions',
    'Type_of_collision', 'Casualty_class',
    'Age_band_of_casualty', 'Cause_of_accident', 'Vehicle_movement'  # âœ… comma added
], axis=1, inplace=True)

df3.head()
```

```python
x=df3.drop(['Accident_severity'],axis=1)
x.shape
```

```python
x.head()
```

```python
y = df3.iloc[:, 2]  # Accesses the third column (index 2)
y.head()
```

```python
#checking the count of eac item in the output column
y.value_counts()
```

```python
#plotting count plot using seaborn
sns.countplot(x = y, palette= 'muted')
```

```python
from imblearn.over_sampling import SMOTE
oversample = SMOTE()
xo, yo = oversample.fit_resample(x, y)
```

```python
#checking the oversampling output
y1 = pd.DataFrame(yo)
y1.value_counts()
```

```python
sns.countplot(x = yo, palette='muted')
```

```python
#converting data to training data and testing data
from sklearn.model_selection import train_test_split

#splitting 70% of the data to traing data and 30% of data to testing data
x_train,x_test,y_train,y_test=train_test_split(xo,yo,test_size=0.30,random_state=42)
```

```python
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
 #KNN model alg
from sklearn.neighbors import KNeighborsClassifier
model_KNN=KNeighborsClassifier(n_neighbors=5)
model_KNN.fit(x_train,y_train)
```

```python
y_pred=model_KNN.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, ConfusionMatrixDisplay
```

```python
report_KNN=classification_report(y_test,y_pred)
print(report_KNN)
```

```python
accuracy_KNN = accuracy_score(y_test, y_pred)
print(accuracy_KNN)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Compute and print matrix
matrix_KNN = confusion_matrix(y_test, y_pred)
print(matrix_KNN, '\n')

# Display the matrix visually
ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
plt.title("Confusion Matrix - KNN")
plt.show()
```
